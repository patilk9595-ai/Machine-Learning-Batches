<h2>K-fold Cross validation</h2>

K-Fold cross validation is a reasampling technique used to evaluate how well a ml model generalizes to unseen data.

Instead of splitting data once into train and test sets, k-fold repeatedly creates different train-test combinations and averages the results.

<h2>How K-fold Works</h2>

Dataset size = 100
K = 5

The dataset is divided into 5 equal parts(folds):

Fold1 = 20 rows
Fold2 = 20 rows
Fold3 = 20 rows
Fold4 = 20 rows
Fold5 = 20 rows


<h3>Iteration 1</h3>

Test = fold1<br>
Train = fold2 + fold3 + fold4 + fold5

<h3>Iteration2</h3>

Test = fold2<br>
Train = fold1 + fold3 + fold4 + fold5

<h3>Iteration3</h3>

Test = fold3<br>
Train = fold1 + fold2 + fold4 + fold5

<h3>Iteration4</h3>

Test = fold4<br>
Train = fold1 + fold3 + fold2 + fold5


<h3>Iteration5</h3>

Test = fold5<br>
Train = fold1 + fold3 + fold4 + fold2



<h2>Final Score</h2>
fold 1 = 91%<br>
fold 2 = 93%<br>
fold 3 = 90%<br>
fold 4 = 94%<br>
fold 5 = 92%<br>

Final score would be average of accuracies

```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

df = pd.DataFrame(load_iris().data,columns=load_iris().feature_names)
df['Target'] = load_iris().target
```

```python
len(df)
```

```python
X = df.drop('Target',axis=1)
y = df['Target']
```

```python
df.tail(10)
```

```python
model = DecisionTreeClassifier(random_state=42)

score = cross_val_score(
    model,
    X,
    y,
    cv=5
)

print(score)
```

```python
print(score.mean())
```
