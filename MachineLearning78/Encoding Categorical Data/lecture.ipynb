<h1>Encoding</h1>

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "StudentID":np.arange(1,101),
    "Subject":np.random.choice(["Math","DSA","React","Chemistry"],100),
    "Marks":np.random.randint(20,50,100),
    "Education":np.random.choice(['High School','Intermediate','Graduate','Post Graduate'],100)
   
}
df = pd.DataFrame(data)
```

```python
df.head()
```

<h2>Label Encoder</h2>

```python
pip install scikit-learn
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()

df['Subject_Trasformed'] = le.fit_transform(df['Subject'])
```

```python
df.head(10)
```

<h2>One Hot Encoding</h2>

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oe = OneHotEncoder()

temp = oe.fit_transform(df[['Subject']]).toarray()
```

```python
df[["Chemistry",'DSA','Math',"React"]] = temp
```

```python
df.head(20)
```

<h2>Ordinal Encoder</h2>

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
categories = [["High School","Intermediate","Graduate","Post Graduate"]]


oen = OrdinalEncoder(categories = categories)

df['EducationTransformed'] = oen.fit_transform(df[['Education']])
```
