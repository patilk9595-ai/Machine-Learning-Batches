```python
import pandas as pd
import numpy as np
data = {
    'Name':['Rohan','Krish','Bob','Sohit','Kashish','Mayank'],
    'Age':[24,np.nan,34,23,np.nan,40],
    'Attendance':[100,90,80,np.nan,90,85]
}

df = pd.DataFrame(data)
```

```python
df['Notes'] = [np.nan,"Good",np.nan,np.nan,np.nan,"Average"]
```

```python
df
```

```python
df['Age'].mean()
```

```python
df['Age_filled'] = df['Age'].fillna(df['Age'].mean())
```

```python
df
```

```python
df['Attendance_filled'] = df['Attendance'].fillna(df['Attendance'].median())
```

```python
df
```

```python
df['Notes_filled'] = df['Notes'].fillna("Not Given")
```

```python
df
```

```python
df['Age_ffill'] = df['Age'].ffill()
```

```python
df
```

```python
df['Attendance_bfill'] = df['Attendance'].bfill()
```

```python
df['Notes_bfill'] = df['Notes'].bfill()
```
