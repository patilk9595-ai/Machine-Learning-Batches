<h2>Types of missing values</h2>

- MCAR - Missing comletely at random
- MAR - Missing at random
- MNAR - Missing not at random

<h3>MCAR</h3>

The probability of a value being missed has absolutly no relationship with any variable in dataset.

<h3>MAR<h3>

The missingness is related to observed data, but not the missing values themseleves.

<h3>MNAR</h3>

The missingness depends on itself.

<h2>Handling Missing Values Methods</h2>

1. Removing Missing values

    - Row Deletion
    - Column Deletion

2. Imputation

    - Univariate Imputation
        1. Numerical Value Imputation
        2. Categorical Value Imputation
    - Multivariate Imputation
        1. KNN-Imputation
        2. Iterative Imputation


```python
import pandas as pd
import numpy as np
data = {
    'Name':['Rohan','Krish','Bob','Sohit','Kashish','Mayank'],
    'Age':[24,np.nan,34,23,np.nan,40],
    'Attendance':[100,90,80,np.nan,90,85]
}

df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()/len(df)*100
```

<h3>Row deletion</h3>

<strong>When it is good to use</strong>

- Low percentage: if < 5% data is missing
- MCAR: The missingness has no hidden pattern
- Large Dataset: if you have large dataset, e.g: if we have 1 million rows, losing 1000 rows won't hurt

<strong>When not to use it</strong>

- Small dataset
- Bias risk: If only "High income" individuals refused to answer the survey, dropping those rows removes your "high income" representation entirely
