```python
def factorial(num):
    fact = 1
    for i in range(1,num+1):
        fact = fact * i
    
    print(fact)
```

```python
def isArmstrong(num):
    pwr = 0
    t = num
    while t>0:
        t = t//10
        pwr = pwr + 1

    #pwr = len(num)
    n = int(num)
    temp = n
    sum = 0
    while n>0:
        re = n%10
        sum =sum + re**pwr
        n = n // 10

    if sum == temp:
        print(temp," is armstrong number")
    else:
        print(temp," is not armstrong number")
```
