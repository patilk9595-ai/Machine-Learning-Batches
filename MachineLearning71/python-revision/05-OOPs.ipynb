#### Encapsulation

```python
class Car:
    tyres = 4
    
    def __init__(self,name,color,en_no):#constructor
        self.name = name#public
        self._color = color#proteced : You can access through object of class
        self.__engine_no = en_no #private variable
        
    def start(self):
        print(self.name,"is starting")
        
    def stop(self):
        self.__check_engine_no()
        print(self.name,"is stopping")
        
    def __check_engine_no(self):
        print("Engine No.:",self.__engine_no) ## you can access the private methods and variables within the class only.
```

```python
c = Car("Xl6","Black","34jkl23kn43")
```

```python
c.name
```

```python
c._color
```

```python
c.__engine_no
```

```python
c.stop()
```

#### Inheritance

```python
class Father:   
    def __init__(self):
        self.sirname="Gupta"
        print("This is father class")
        
    def drive(self):
        print("I can drive")
        
class Child(Father):   
    def __init__(self):#initialized
        print("This is child class")
        
    def study(self):
        print("I do study")
```

```python
c = Child()
```

```python
f = Father()
```

```python
c.drive()
```

```python
f.study()
```

```python
c.sirname
```

#### Polymorphism

```python
class Shape:#Parent
    def area(self):
        print("This method will give you area of shape")
        
class Rectangle(Shape):
    def __init__(self,l,w):
        self.l = l
        self.w = w
        
    def area(self):
        print("Area of Rectangle: ",self.l*self.w)
        
class Square(Shape):
    def area(self,e):
        print("Area of square: ",e*e)
        
class Circle(Shape):
    def area(self,r):
        print("Area of circle: ",3.14* (r**2))
```

```python
s = Shape()
r = Rectangle(12,23)
sq = Square()
c = Circle()
```

```python
s.area()
```

```python
r.area()
```

```python
sq.area(12)
```

```python
c.area(23)
```

### Abstraction

```python
from abc import ABC,abstractmethod
```

```python
class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass
```

```python
class Car(Vehicle):
    def start(self):
        print("Car is starting")
        
class Bike(Vehicle):
    def start(self):
        print("Bike is starting")
```

```python
c = Car()
b = Bike()
```

```python
c.start()
```

```python
b.start()
```
