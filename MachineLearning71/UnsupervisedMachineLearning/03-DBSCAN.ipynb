<h1>DBSCAN(Density-Based Spatial clustering of Applications with Noise)</h1>

DBSCAN is one of the most powerful clustering algorithms in machine learning. Unlike K-Means, it does not require specifying the number of clusters beforehand and can identify outliers (noise points) automatically.

<h3>Parameters</h3>

1. Epsilon (Radius around a point)
2. MinPts (Minimum points required inside epsilon-neighborhood)

<h3>Types of Points</h3>

<h4>Core Point</h4>

A point having at least MinPts points inside epsilon radius.

<h4>Border Point</h4>

Not enough neighbors itself but reachable form a core point

<h4>Noise Point</h4>

Neigther core nor border, not wrapping other points, they isolated

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
X , y = make_moons(
    n_samples=500,
    noise=0.15,
    random_state=42

)

print(X.shape)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X[:,0],X[:,1],c=y)
plt.title("Original dataset")
plt.show()
```

```python
dbscan = DBSCAN(
    eps=0.15,
    min_samples=5
)

labels = dbscan.fit_predict(X)
```

```python
np.unique(labels)
```

```python
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

n_clusters
```

```python
list(labels).count(1)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X[:,0],X[:,1],c=labels,cmap='rainbow')
plt.title("DBSCAN Clustering")
plt.show()
```
